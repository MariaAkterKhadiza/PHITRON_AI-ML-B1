## Explain the concept of backpropagation in detail. Describe how the error is propagated from the output layer to the hidden layers. Derive the mathematical expressions for weight updates using the chain rule, and explain how learning rate affects the training process.


Write Answer 01:


Backpropagation is a learning technique used in neural networks to reduce errors. At first, the network receives input data and moves it forward through several layers to generate an output. This predicted output is then compared with the actual target value to determine the error.
After that, the process moves in reverse—from the output layer back to the hidden layers—to identify where the mistakes occurred. During this step, the network measures how much each neuron and connection (weight) contributed to the overall error. This is done using the chain rule, which breaks a complex calculation into smaller parts to see how each component influences the final result.

Once these contributions are known, the weights are adjusted using an
 update

 rule:

w(new) = w(old) − η × (∂Loss / ∂Weight)


Here, the learning rate (η) controls how large the updates are. If the learning rate is very small, the training process becomes slow. On the other hand, if it is too large, the model may become unstable and fail to learn effectively.


In short, backpropagation allows the neural network to learn by calculating errors, propagating them backward, and gradually updating weights to improve prediction accuracy.

# Question 03: [Marks 5]

## Discuss the vanishing gradient problems in deep neural networks. Explain why they occur during backpropagation, how activation functions contribute to this issue, and suggest at least three techniques to overcome these problems.


## Write Answer 03:

The vanishing gradient problem appears in deep neural networks when the gradients become very small as they travel backward through multiple layers. In backpropagation, gradients are computed using the chain rule, which requires multiplying several derivatives together. When these derivatives are less than 1, repeated multiplication causes the gradient to decrease rapidly as it moves toward the earlier layers. Because of this, the weights in the first few layers get extremely tiny updates, making learning very slow or sometimes stopping it altogether.

Activation functions such as sigmoid (and sometimes even ReLU in certain cases) can contribute to this issue because they compress values into a limited range. This leads to small derivative values. During backpropagation, these small derivatives are multiplied across many layers, which further reduces the gradient. As a result, the earlier layers fail to learn effectively, and the overall performance of the network suffers.

There are several ways to address this problem, including:

Simplifying the network architecture (reducing depth or complexity)
Using Batch Normalization to stabilize and scale activations
Applying ReLU or its variants to maintain stronger gradients

# Question 04: [ Marks 5 ]

## Describe the architecture of an Artificial Neural Network (ANN) in detail. Explain the role of each component (neurons, weights, bias, activation functions). Also discuss how increasing the number of hidden layers and neurons affects model performance and overfitting.


## Write Answer 04:

An Artificial Neural Network (ANN) is a computational model composed of multiple layers of interconnected neurons that process information in a step-by-step manner. It consists of three main components: an input layer that receives feature values, one or more hidden layers that extract and learn patterns from the data, and an output layer that produces the final prediction.

Inside each neuron, the input values are multiplied by their corresponding weights, a bias term is added, and the result is passed through an activation function such as ReLU or sigmoid to introduce non-linearity into the model. In mathematical form, a neuron can be expressed as:

z = W · x + b
a = f(z)

where W represents the weights, b is the bias, and f denotes the activation function.

As the number of hidden layers and neurons increases, the network gains the ability to learn more complex and abstract patterns, which can improve performance. However, if the model becomes too complex, it may lead to overfitting—where the network performs well on training data but poorly on unseen data. Therefore, it is important to maintain a balance between model complexity and generalization ability.

## Compare Artificial Neural Networks (ANN) and Convolutional Neural Networks (CNN) in detail. Discuss their architectures, working principles, advantages, limitations, and typical use cases such as image processing, tabular data, and pattern recognition.

## Write Answer 05:

Artificial Neural Networks (ANNs) and Convolutional Neural Networks (CNNs) are both types of deep learning models, but they are used for different kinds of data and applications. An ANN has a straightforward structure made up of fully connected layers, where each neuron is linked to every neuron in the next layer. It processes inputs using the equation z = W · x + b and applies an activation function to learn patterns. ANNs are mainly suitable for structured or tabular data, but they are not effective at capturing spatial relationships within data.

On the other hand, CNNs are specially designed for image and grid-like data. They use convolutional layers with filters to automatically detect important features, and pooling layers to reduce the size of the data while keeping essential information. This makes CNNs more efficient and powerful for visual and spatial tasks.

Working principle:

ANN: Learns patterns by considering all input features together
CNN: First learns local patterns (like edges or shapes), then combines them to form more complex features

Advantages:

ANN:

Simple architecture and easy to implement
Works well for tabular or structured datasets

CNN:

Very effective for image and visual data
Uses parameter sharing, so it has fewer parameters
Automatically learns important features without manual extraction

Limitations:

ANN:

Not suitable for image data
Large number of parameters can lead to overfitting

CNN:

Requires higher computational power
Needs large datasets and careful tuning

Applications:

ANN:

Tabular data like finance or medical records
Basic classification and regression tasks

CNN:

Image recognition, face detection, and object detection
Computer vision and pattern recognition tasks

## Explain the concept of pooling in CNNs. Compare MaxPooling and AveragePooling in terms of operation, advantages, and impact on feature extraction. Discuss how pooling helps in reducing overfitting and computational complexity.

Write Answer 08:

Comparison: Max Pooling vs Average Pooling

Max Pooling selects the highest value from a specific region of the feature map. This means that if a strong feature such as an edge, corner, or bright spot exists within that area, it will be preserved. Because of this behavior, Max Pooling is very effective for detecting important features in images, especially when the exact location of the feature is not very important.

Average Pooling, in contrast, calculates the average of all values within the selected region. Instead of focusing on the most prominent feature, it provides a general representation of the area. This results in smoother outputs, but it may lose sharp or detailed information. Therefore, it is less powerful than Max Pooling for capturing strong or distinctive patterns.

Why Pooling is Important

Reduces Overfitting:
Pooling decreases the spatial size of feature maps, which reduces the number of parameters in the model. With fewer details, the network is less likely to memorize noise from the training data. Instead, it focuses on learning more general patterns, which improves performance on unseen data.

Reduces Computational Cost:
By shrinking feature maps (for example, from 28×28 to 14×14), pooling reduces the amount of computation required in later layers. This leads to:

Faster training
Lower memory usage
More efficient model performance

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
from torch.utils.data import Dataset,DataLoader
import torch.nn as nn
import torch.optim as optim

torch.manual_seed(42)
```

```python
# check the gpu
device= torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device :{device}")
```

```python
df=pd.read_csv("fmnist_small.csv")
df.head(5)
```

```python
# Create a 4x4 grid of images
fig, axes = plt.subplots(4, 4, figsize=(10, 10))
fig.suptitle("First 16 Images", fontsize=16)

# Plot the first 16 images from the dataset
for i, ax in enumerate(axes.flat):
    img = df.iloc[i, 1:].values.reshape(28, 28)  # Reshape to 28x28
    ax.imshow(img)  # Display in grayscale
    ax.axis('off')  # Remove axis for a cleaner look
    ax.set_title(f"Label: {df.iloc[i, 0]}")  # Show the label

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to fit the title
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('label',axis=1)
y=df['label']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train=x_train/255.0
x_test=x_test/255.0
```

```python
class CustomDataset(Dataset):
  def __init__(self,features,labels):
    # convert to pytorch to tensor
    self.features=torch.tensor(features.values,dtype=torch.float32).reshape(-1,1,28,28)
    self.labels=torch.tensor(labels.values,dtype=torch.long)
  def __len__(self):
    return len(self.features)

  def __getitem__(self, index):
     return self.features[index],self.labels[index]
```

```python
train_dataset=CustomDataset(x_train,y_train)
test_dataset =CustomDataset(x_test,y_test)
```

```python
train_loader= DataLoader(train_dataset,batch_size=32,shuffle=True,pin_memory=True)
test_loader=DataLoader(test_dataset,batch_size=32,shuffle=True,pin_memory=True)
```

```python
class MyNN(nn.Module):
    def __init__(self, input_features):
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(input_features, 32, kernel_size=3, padding='same'),
            nn.ReLU(),
            nn.BatchNorm2d(32),
            nn.MaxPool2d(kernel_size=2, stride=2),

            nn.Conv2d(32, 64, kernel_size=3, padding='same'),
            nn.ReLU(),
            nn.BatchNorm2d(64),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64*7*7, 128),
            nn.ReLU(),
            nn.Dropout(p=0.4),

            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(p=0.4),

            nn.Linear(64, 10)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)

        return x


learning_rate = 0.01
epochs = 100
model = MyNN(1)

model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=learning_rate, weight_decay=1e-4)
```

```python
for epoch in range(epochs):

  total_epoch_loss = 0

  for batch_features, batch_labels in train_loader:

    # move data to gpu
    batch_features, batch_labels = batch_features.to(device), batch_labels.to(device)

    # forward pass
    outputs = model(batch_features)

    # calculate loss
    loss = criterion(outputs, batch_labels)

    # back pass
    optimizer.zero_grad()
    loss.backward()

    # update grads
    optimizer.step()

    total_epoch_loss = total_epoch_loss + loss.item()

  avg_loss = total_epoch_loss/len(train_loader)
  print(f'Epoch: {epoch + 1} , Loss: {avg_loss}')
```

```python
model.eval()
```

```python
total = 0
correct = 0

with torch.no_grad():

  for batch_features, batch_labels in test_loader:

    # move data to gpu
    batch_features, batch_labels = batch_features.to(device), batch_labels.to(device)

    outputs = model(batch_features)

    _, predicted = torch.max(outputs, 1)

    total = total + batch_labels.shape[0]

    correct = correct + (predicted == batch_labels).sum().item()

print(correct/total)
```
