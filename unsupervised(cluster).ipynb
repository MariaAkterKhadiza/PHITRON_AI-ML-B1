```python
from sklearn.cluster import KMeans
import numpy as np

# Input data (study hours)
X = np.array([[1], [2], [3], [8], [9], [10]])

# Create model (2 groups)
model = KMeans(n_clusters=2)

# Train model
model.fit(X)

# Predict groups
labels = model.predict(X)

print("Groups:", labels)
```
