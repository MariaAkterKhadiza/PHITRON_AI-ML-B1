**classification**

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

# Input (study hours)
X = np.array([1, 2, 3, 4]).reshape(-1, 1)

# Output (0 = Fail, 1 = Pass)
y = np.array([0, 0, 1, 1])

# Create model
model = LogisticRegression()

# Train model
model.fit(X, y)

# Predict
print("Pass/Fail for 3 hours:", model.predict([[3]]))
```

**regression **

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Input (study hours)
X = np.array([1, 2, 3, 4]).reshape(-1, 1)

# Output (marks)
y = np.array([40, 50, 60, 70])

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict
print("Marks for 5 hours:", model.predict([[5]]))
```
