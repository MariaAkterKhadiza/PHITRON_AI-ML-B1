```python
# nd declaration
import numpy as np

# 1D array
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1.ndim)  # Output: 1 (1 dimension)

# 2D array
arr2 = np.array([
    [1, 2, 3, 4],  # row1
    [6, 7, 8, 9]   # row2
])
print(arr2.ndim)  # Output: 2 (2 dimensions)

# 3D array
arr3 = np.array([
    # first floor
    [
        [1, 2, 3, 4],  # row1
        [6, 7, 8, 9]   # row2
    ],
    # second floor
    [
        [1, 2, 3, 4],  # row1
        [6, 7, 8, 9]   # row2
    ]
])
print(arr3.ndim)  # Output: 3 (3 dimensions)
```

```python
#array attributes

#dimension
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)


#shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)


#data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

#size
print(arr1.size)
print(arr2.size)
print(arr3.size)
```

```python
#data type

arr9 = np.array([1,2,3,4,5])
print(arr9.dtype)

arr9 = np.array([1.5,2.5,3.5,4.5,5.5])
print(arr9.dtype)

arr9 = np.array([1,2,3,4,5.5])
#upcasting
print(arr9.dtype)

arr9 = np.array([1,2,3,4,'3'])
print(arr9.dtype)

```
