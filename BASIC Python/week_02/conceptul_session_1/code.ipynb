function

```python
def greet(name):
    print("hello",name)
greet("maria")
```

```python
#positional parameter
def greet(name2,name1):
    print("hello",name1,name2)
greet("maria","akash")
```

```python
#keyword 
def info(name,age):
    print(f"{name} age is:{age}")
```

```python
def add(a,b,c,d):
    return a + b + c + d
print(add(1,2,3,4))
```

```python
def add(*a):
  print(a)
(add(1,2,3,4))
```

```python
def info(**k):
    print(type(k))
    print(f"{k["name"]} age is :{k["age"]} country :{k["country"]}")
info(age=20, name="maria",country="bd")
```
