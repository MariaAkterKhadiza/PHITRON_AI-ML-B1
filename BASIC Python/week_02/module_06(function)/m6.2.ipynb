```python

def square_addition(a,b,c):
    print(f"{a} {b} {c}")
    a = a**2
    b = b**2
    c = c**2
    return a+b+c
ans = square_addition(10,20,30)
print(ans)
```

```python
def square_addition(*args):
    summation = 0
    for i in args:
        i = i ** 2
        summation += i
    return summation

ans = square_addition(10, 20, 30, 40, 50, 90)
print(ans)
```

```python
#generator

lst = [x for x in range(500)]  # data set

def data_loaded(chunk_size, lst):
    for i in range(0, len(lst), chunk_size):
        yield lst[i:i+chunk_size]

x = data_loaded(5, lst)

print(next(x))
print(next(x))
```
