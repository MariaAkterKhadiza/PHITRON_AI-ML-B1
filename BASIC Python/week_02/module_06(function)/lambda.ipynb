```python
def sq_add(x,y):
    return x**2 + y**2
result = sq_add(2,3)
print(result)


#lambda
sq_add2 = lambda x,y : x**2 + y**2
print(sq_add2(2,3))

#
square = lambda n : n**2
print(square(4))

#
add = lambda a,b : a +b
print(add(5,10))

#
greet = lambda name :f"hello {name}"
print(greet("Adil"))

#
check_even = lambda x :"even" if x%2 == 0 else "Odd"
print(check_even(5))

#function
result = (lambda x,y : x*y)(5,3)
print(result)

```
