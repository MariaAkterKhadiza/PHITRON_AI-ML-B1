```python
#class & object

class Phone:
    category = "Electronis" #attribute

    #constructor

    def __init__(self,model , battery ,  camera , battery_percentange = 100):
        self.model = model
        self.battery = battery
        self.camera = camera
        self.battery_percentange = battery_percentange 

    #method
    def charge(self,hours):
        self.charge += hours
```

```python
##create object
##class attribute
apple = Phone()
apple.category = "super elecronis"
bluebery = Phone()

motorola = Phone()


print(apple.category)
print(bluebery.category)
print(motorola.category)
print(motorola.battery_percentange)
```

```python
#instance attribute
apple = Phone("iphone17",3000,40)

bluebery = Phone("B_17",4000,30)

motorola = Phone("m-17",3500,35)

print(motorola.camera)
```
