```python
class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return 3.1416 * self.r * self.r


rect = Rectangle(5, 10)
circle = Circle(7)

print(f"Rectangle Area: {rect.area():.2f}")
print(f"Circle Area: {circle.area():.2f}")
```
